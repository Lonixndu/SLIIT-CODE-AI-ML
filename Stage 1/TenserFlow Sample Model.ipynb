# Import data

A raw dataset is imported using pandas package.


```python
import pandas as pd
url = "https://raw.githubusercontent.com/selva86/datasets/master/Ionosphere.csv"
df = pd.read_csv(url)
df.head(5)
```

X axis represents the independent variables; which are 'V1' to 'V34'
Y axis represents the dependent variables; which is 'Class'

```python
x = df.drop(['Class'], axis = 1)
y = df['Class']
print(y)
```

Then the both axis will be split into two parts; one for "Train", other for "Test".
Because first, we should train the model with data, and test them with another set of data.

Here, 80% of data will be used for training, and the remaining 20% will be used for testing

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size = 0.2)
#0.2 means 20% from the dataset
```

# Import Dependencies

```python
from keras.models import Sequential, load_model
from keras.layers import Dense, Dropout
```

# Build Model
In ANN, there are multiple hidden layers stacked on each other, to give an output.
in order to stack these layers one under the other; the Sequential method is used.



*   No of neurons will be defined as
  'units'.
*   No of inputs will be defined as 'input dimensions'



**First hidden layer;**
  10 neurons, **relu** activation function is used, data cominng with  minus values will automatically assigned to 0.

**Second hidden layer;**
  5 neurons, same activation function as the first layer

**Output Layer;**
  1 neuron (returns 1 value; 1=true or 0=false), **sigmoid** activation funcion is used, returns a probability of a value.
  

```python
model = Sequential()
model.add(Dense(units=10, activation='relu', input_dim=len(xTrain.columns)))
# x train columns;  independent variable columns. for the csv imported, there are 34.
model.add(Dense(units=5, activation='relu',))
model.add(Dense(units=1, activation='sigmoid'))

model.summary()
```

```python
#compile the model
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

epochs = how many times whole dataset should be trained to the model again and again.

batch_size = total number of training samples in a single batch.

in the output the number of batches needs to be completed in one epoch is known as the iterations. In the dataset, the rows/training data divided by the batch_size, will give the number of iterations. (for this example it is 9)

```python
#create the model
model.fit(xTrain, yTrain, validation_data=(xTest, yTest), epochs=100, batch_size=32)
```

```python
model.save('model.h5')
```
